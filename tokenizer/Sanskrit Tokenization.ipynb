```python
import sentencepiece as spm
import pickle
import pathlib
```

```python
path = pathlib.Path('/home/gaurav/PycharmProjects/nlp-for-sanskrit/language-model')
```

```python
p = path.glob('SanskritWikipediaArticles/*')
files = [x for x in p if x.is_file()]
```

```python
files = [str(file) for file in files]
```

```python
flist = ','.join(files)
```

```python
spm.SentencePieceTrainer.Train(f'--input={flist} --model_prefix=sanskrit_lm --vocab_size=20000')
```
